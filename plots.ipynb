```python
from matplotlib.pyplot import figure
from matplotlib import pyplot as plt
import seaborn as sns
import cv2
```

To plot Loss Function MSE vs Epochs

```python
### The following function shows the first n sample images

def images_show(n,X):
  plt.figure(figsize=(30,50))
  for i in range(0,n) :
      plt.subplot(10,5,i+1)
      plt.axis('off')
      plt.imshow(X[i])
```

```python
def plot_loss(train,model) :
    accuracy = train.history['loss']
    val_accuracy = train.history['val_loss']
    epochs = range(len(accuracy))
    figure(figsize=(8, 6), dpi=80)
    plt.plot(epochs, accuracy, 'g', label='Training Set')
    plt.plot(epochs, val_accuracy, 'r', label='Testing Set')
    plt.title(model+' - Loss Function Value vs Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('MSE Loss Function Value')
    plt.legend()
    plt.show()
```

To plot MAE Mean Absolute Error vs Epochs

```python
def plot_scores(train,model) :
    accuracy = train.history['mean_absolute_error']
    val_accuracy = train.history['val_mean_absolute_error']
    epochs = range(len(accuracy))
    figure(figsize=(8, 6), dpi=80)
    plt.plot(epochs, accuracy, 'g', label='Training Set')
    plt.plot(epochs, val_accuracy, 'r', label='Testing Set')
    plt.title(model+' - MAE vs Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Mean Absolute Error')
    plt.legend()
    plt.show()
```

To plot the boundaries for first n images based on prediction

```python
def boundaries(n,y_cnn,y_test,X_test,output_folder):
  plt.figure(figsize=(20,40))
  for i in range(0,n) :
      plt.subplot(10,5,i+1)
      plt.axis('off')
      ny = y_cnn[i]*255
      nt = y_test[i]*255
      ### Adding original license plate boundary
      image = cv2.rectangle(X_test[i],(int(ny[0]),int(ny[1])),(int(ny[2]),int(ny[3])),(0, 255, 0))
      ### Adding predicted license plate boundary
      cv2.rectangle(image,(int(nt[0]),int(nt[1])),(int(nt[2]),int(nt[3])),(255,0, 0))
      #plt.imsave(output_folder+"/image_"+str(i)+'.png', image)
      plt.imshow(image)
```
